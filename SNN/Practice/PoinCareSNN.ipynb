```julia
using PyPlot, StaticArrays
```

```julia
# Based on the great tutorial by Rainer Engelken: https://www.youtube.com/watch?v=cLLQcshWEbE
function snnPhase(topology, nSpikes)    
    c = 1.0   
end    

function spikingNet(topology)    
    nSpikes = 10^7    
    topo = SMatrix{3, 3}(0.<topology)    
    c = 1.0    
    ϕ = rand(3)
    Ψ = Float64[]    
    tic()
    for s = 1:nSpikes        
        ϕmax, j = findmax(ϕ)
        dt = π/2 - ϕmax        
        ϕ += dt
        post = topo[:,j]        
        ϕ[post] = atan.(tan.(ϕ[post]) - c)
        ϕ[j] = -π/2        
        j == 1 && append!(Ψ, ϕ[2:3])        
    end
    toc()

    plot(Ψ[1:2:end], Ψ[2:2:end], ",k", ms = 0.01, alpha = 0.01); axis("off")
    nothing
end

function spikingNetPhase(topology)    
    nSpikes = 10^2
    topo = SMatrix{3, 3}(0.<topology)
    c = 1.0    
    ϕ = rand(3)
    Ψ = Float64[]    
    tic()
    for s = 1:nSpikes        
        ϕmax, j = findmax(ϕ)
        dt = π/2 - ϕmax        
        ϕ += dt
        post = topo[:,j]        
        ϕ[post] = atan.(tan.(ϕ[post]) - c)        
        ϕ[j] = -π/2        
        j == 1 && append!(Ψ, ϕ[2:3])        
    end
    toc()
    
    plot(Ψ, ",k", ms = 0.01, alpha = 0.01); axis("off")
    print(Ψ)
    nothing
end    
```
